```python
# Create a dictionary to map the landmarks to the numbers according to the MediaPipe Holistic Model above
landmark_dict = {
    'NOSE': 0,
    'LEFT_EYE_INNER': 1,
    'LEFT_EYE': 2,
    'LEFT_EYE_OUTER': 3,
    'RIGHT_EYE_INNER': 4,
    'RIGHT_EYE': 5,
    'RIGHT_EYE_OUTER': 6,
    'LEFT_EAR': 7,
    'RIGHT_EAR': 8,
    'MOUTH_LEFT': 9,
    'MOUTH_RIGHT': 10,
    'LEFT_SHOULDER': 11,
    'RIGHT_SHOULDER': 12,
    'LEFT_ELBOW': 13,
    'RIGHT_ELBOW': 14,
    'LEFT_WRIST': 15,
    'RIGHT_WRIST': 16,
    'LEFT_PINKY': 17,
    'RIGHT_PINKY': 18,
    'LEFT_INDEX': 19,
    'RIGHT_INDEX': 20,
    'LEFT_THUMB': 21,
    'RIGHT_THUMB': 22,
    'LEFT_HIP': 23,
    'RIGHT_HIP': 24,
    'LEFT_KNEE': 25,
    'RIGHT_KNEE': 26,
    'LEFT_ANKLE': 27,
    'RIGHT_ANKLE': 28,
    'LEFT_HEEL': 29,
    'RIGHT_HEEL': 30,
    'LEFT_FOOT_INDEX': 31,
    'RIGHT_FOOT_INDEX': 32
}

key_list = list(landmark_dict.keys())
```

**Note:** The landmark coordinates in the MediaPipe Holistic Model begin from '0' while the coordinates captured in the csv begin from '1', we need to '-1' from the numeric part of the coordinates captured before matching the landmark names in the MediaPipe Holistic Model.

The landmark names will remain in uppercase to be identified more easily.

```python
# Define a function that transforms the numeric part of the coordinates captured to the landmark names
def num_to_name(df):
    
    # Create an empty dictionary to store the new column names
    column_mapping = {}

    for column in df.columns:
        # Create a condition that if all column headers contain 'x', 'y', 'z' or 'v' in the first letter
        if column[0] in ('x', 'y', 'z', 'v'):
            
            # Extract the numeric part
            numeric_part = int(''.join(filter(str.isdigit, column)))
            
            # Minus 1 from the numeric part to match the index of the MediaPipe Holistic Model
            key_index = numeric_part - 1
            
            # Map the numeric parts to the key index
            if 0 <= key_index < len(key_list):
                new_name = f'{column[0]}_{key_list[key_index]}'
                column_mapping[column] = new_name

    # Rename the columns
    df.rename(columns=column_mapping, inplace=True) 
    return df
```

```python
# Transform all numeric part of the column headers to landmark names by applying the above function
reg_coord = num_to_name(reg_coord)
pulse_coord = num_to_name(pulse_coord)
push_coord = num_to_name(push_coord)
sc_coord = num_to_name(sc_coord)

# Display transformed coordinates dfs
print('reg_coord')
display(reg_coord.head(3))
print()
print('pulse_coord')
display(pulse_coord.head(3))
print()
print('push_coord')
display(push_coord.head(3))
print()
print('sc_coord')
display(sc_coord.head(3))
print()
```

### `prompts`
In a spin class, the instructor prompts the students on what choreography moves to perform at different parts of the song. To simulate this, the game would prompt the rider on the moves and predict whether they have done the moves correctly.

In each dataframe, I would be creating a new column displaying the prompted moves. Since the application has not been developed and the videos were recorded without any prompts, this column would be filled in manually. 

Ultimately, we want to compare the values of `choreo` and `prompts`; if they match, the rider has performed the choreography correctly and vice versa. Another column, `choreo_result` would be created subsequently to show the result in binary form - 0 for correct moves and 1 for incorrect moves. 

To prevent class imbalance, the first half of the `prompts` column would be filled with the correct `choreo`, which is named the `main_choreo`. The remaining half would be filled with the other choreography moves one after another until the end of the column. This should produce an even split of 0 and 1 in the `choreo_results` column.

```python
# Define a function to create the 'prompts' column
def prompted_moves(df, main_choreo, choreo2, choreo3, choreo4):
    
    # Find total number of rows 
    num_rows = len(df)
    
    # Find the number of rows to be filled by the main_choreo
    num_main_choreo = int(num_rows // 2)
    
    # Find the remaining number of rows to be filled by the other choreos
    num_other = num_rows - num_main_choreo
    
    # Fill the first half with the main_choreo
    main_list = [main_choreo] * num_main_choreo
    
    # Create an empty list to store the remaining choreos
    other_list = []
    
    # List the other choreos
    other_choreos = [choreo2, choreo3, choreo4]
    
    # Append the remaining half with the other choreos one after another until the final row
    while len(other_list) < num_other:
        for choreo in other_choreos:
            other_list.append(choreo)
            
            if len(other_list) == num_other:
                break
                
    # Combine the two lists into the new column, 'prompts'
    df['prompts'] = main_list + other_list
    
    return df
```

```python
# Apply the above function to all dataframes
reg_coord = prompted_moves(reg_coord, 'regular', 'pulse', 'push', 'sexy_corners')
pulse_coord = prompted_moves(pulse_coord, 'pulse', 'push', 'sexy_corners', 'regular')
push_coord = prompted_moves(push_coord, 'push', 'sexy_corners', 'regular', 'pulse')
sc_coord = prompted_moves(sc_coord, 'sexy_corners', 'regular', 'pulse', 'push')

# Display
print('reg_coord')
display(reg_coord[['timestamp', 'choreo', 'prompts']].head())
print(reg_coord['prompts'].value_counts())
print()
print('pulse_coord')
display(pulse_coord[['timestamp', 'choreo', 'prompts']].head())
print(pulse_coord['prompts'].value_counts())
print()
print('push_coord')
display(push_coord[['timestamp', 'choreo', 'prompts']].head())
print(push_coord['prompts'].value_counts())
print()
print('sc_coord')
display(sc_coord[['timestamp', 'choreo', 'prompts']].head())
print(sc_coord['prompts'].value_counts())
```

## Cleaning
### Transpose dataframe
In `Part I - Imports`, it was explained that the column headers of a tempogram ratio array represent the different tempo hypotheses (a set of possible tempos for an audio signal), while the rows represent the different time points (marks the start of each window length of an audio signal) in the audio signal (see below for a sample layout).

| Time point | Tempo hypothesis 1 | Tempo hypothesis 2 | Tempo hypothesis 3 |
|---|---|---|---|
| 0 | 1.0 | 0.5 | 0.25 |
| 1 | 0.75 | 0.5 | 0.25 |
| 2 | 0.5 | 0.75 | 0.5 |
| 3 | 0.25 | 1.0 | 0.75 |
| 4 | 0.25 | 0.75 | 1.0 |
<center>(Sample layout of a tempogram)</center><br>

The TGR dataframes will be transposed for better readability.

```python
# Define a function to transpose a dataframe
def transpose(df):
    df = df.T
    return df
```

### Transform time points to timestamps
The time points illustrate the number of frames at the start of each window. To match the time element with the coordinates dataframes, the time points will be converted into timestamp format.

```python
def to_timestamp(y, sr, df):
    
    # Set hop_length to 512, the default
    hop_length=512
    
    # Compute the duration of each frame
    frame_duration = hop_length / sr

    # Create a new list of column headers
    timestamps = []
    for i in range(df.shape[1]):
        timestamps.append(frame_duration * i)
        
    # Round the timestamps up to 2 decimal places
    for i in range(len(timestamps)):
        timestamps[i] = round(timestamps[i], 2)

    # Set the column headers of the DataFrame to the new list of column headers
    df.columns = timestamps

    return df
```

```python
# Transpose all the TGR dataframes and convert time points to timestamps by applying the functions above
reg_tgr = to_timestamp(y_reg, sr_reg, transpose(reg_tgr))
pulse_tgr = to_timestamp(y_pulse, sr_pulse, transpose(pulse_tgr))
push_tgr = to_timestamp(y_push, sr_push, transpose(push_tgr))
sc_tgr = to_timestamp(y_sc, sr_sc, transpose(sc_tgr))

# Display transformed dataframes
print('reg_tgr')
display(reg_tgr.head(3))
print()
print('pulse_tgr')
display(pulse_tgr.head(3))
print()
print('push_tgr')
display(push_tgr.head(3))
print()
print('sc_tgr')
display(sc_tgr.head(3))
```
